```python
from datasets import load_dataset

# Load the dataset
ds = load_dataset("Bilal-Mamji/Medical-summary")

# Inspect the dataset
print(ds)
```

```python
# Check dataset structure
print(ds)

# Check a sample from the train set
print(ds['train'][0])
```

```python

print(ds['train'].column_names)


print(ds['train'][0])
```

```python

def preprocess_data(example):
    dialogue = example['input']
    summary = example['output']
    return {'input_text': dialogue, 'target_text': summary}

processed_dataset = ds.map(preprocess_data)

print(processed_dataset['train'][0])
```

```python
from transformers import AutoTokenizer


tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neo-1.3B")


tokenizer.pad_token = tokenizer.eos_token

def tokenize_function(examples):
    inputs = tokenizer(examples['input_text'], truncation=True, padding='max_length', max_length=512)
    targets = tokenizer(examples['target_text'], truncation=True, padding='max_length', max_length=512)
    inputs['labels'] = targets['input_ids']
    return inputs

tokenized_dataset = processed_dataset.map(tokenize_function, batched=True)


print(tokenized_dataset['train'][0])
```

```python

train_test_split = processed_dataset['train'].train_test_split(test_size=0.1)


train_dataset = train_test_split['train']
test_dataset = train_test_split['test']

print(f"Training examples: {len(train_dataset)}")
print(f"Testing examples: {len(test_dataset)}")

```

```python

train_test_split = tokenized_dataset['train'].train_test_split(test_size=0.1)


train_dataset = train_test_split['train']
test_dataset = train_test_split['test']

print(f"Training examples: {len(train_dataset)}")
print(f"Testing examples: {len(test_dataset)}")
```

```python

train_dataset = train_dataset.remove_columns([col for col in train_dataset.column_names if col not in ['input_ids', 'labels']])
test_dataset = test_dataset.remove_columns([col for col in test_dataset.column_names if col not in ['input_ids', 'labels']])

print("Train dataset columns:", train_dataset.column_names)
print("Test dataset columns:", test_dataset.column_names)
```

```python

print("Train dataset:", train_dataset if 'train_dataset' in locals() else "Not defined")
print("Test dataset:", test_dataset if 'test_dataset' in locals() else "Not defined")
print("Train dataset columns:", train_dataset.column_names if 'train_dataset' in locals() else "No columns")
print("Test dataset columns:", test_dataset.column_names if 'test_dataset' in locals() else "No columns")
```

```python
from transformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained("EleutherAI/gpt-neo-1.3B")
```

```python
from transformers import TrainingArguments


training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=2,
    num_train_epochs=3,
    weight_decay=0.01,
    save_steps=100,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=10,
    fp16=True,
    remove_unused_columns=False
)
```

```python
from transformers import Trainer

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer
)
```

```python

import os
os.environ["WANDB_DISABLED"] = "true"

trainer.train()
```

```python

def generate_soap_summaries(test_dataset, model, tokenizer):
    model.eval()
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)

    predictions = []
    references = []

    for example in test_dataset:

        input_text = example["input_text"]
        inputs = tokenizer(
            input_text, return_tensors="pt", max_length=512, truncation=True, padding="max_length"
        )
        inputs = {key: value.to(device) for key, value in inputs.items()}


        with torch.no_grad():
            outputs = model.generate(**inputs, max_length=512, num_beams=4, early_stopping=True)
        generated_summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

        predictions.append(generated_summary)
        references.append(example["target_text"])  # Ground truth summary

    return predictions, references


test_data = [example for example in test_dataset]
generated_summaries, true_summaries = generate_soap_summaries(test_data, model, tokenizer)


for i in range(5):
    print(f"--- Example {i + 1} ---")
    print("Dialogue:")
    print(test_data[i]["input_text"])
    print("\nGenerated SOAP Summary:")
    print(generated_summaries[i])
    print("\nTrue SOAP Summary:")
    print(true_summaries[i])
    print("\n")
```
